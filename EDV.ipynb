```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score

# Vorbereitung der Feature-Spalten und des Zielvektors
X = df.drop('Price (USD)', axis=1)
y = df['Price (USD)']

# Aufteilung der Daten in Trainings- und Testdaten
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Erstellen eines ColumnTransformers für das One-Hot-Encoding der kategorialen Daten
categorical_features = ['Make', 'Model', 'Fuel Type']

one_hot_encoder = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(transformers=[
    ('cat', one_hot_encoder, categorical_features)
], remainder='passthrough')

# Erstellen und Konfigurieren des Regressionsmodells
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Training des Modells
model.fit(X_train, y_train)

# Vorhersagen auf den Testdaten
y_pred = model.predict(X_test)

# Bewertung des Modells
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```

```python
def predict_car_price(trained_model):
    # Die Schleife läuft weiter, bis der Benutzer "Exit" eingibt
    while True:
        command = input("Enter 'Exit' to quit or press Enter to continue: ")
        if command.lower() == 'exit':
            break
        
        # Eingabeaufforderungen für Benutzer
        make = input("Enter the car's make (e.g., Toyota): ")
        model_name = input("Enter the car's model (e.g., Camry): ")
        year = int(input("Enter the car's year (e.g., 2022): "))
        engine_size = float(input("Enter the engine size in liters (e.g., 2.5): "))
        fuel_type = input("Enter the fuel type (e.g., Petrol): ")

        # Neuen Datenrahmen aus Benutzereingaben erstellen
        new_data = pd.DataFrame({
            'Make': [make],
            'Model': [model_name],
            'Year': [year],
            'Engine Size (L)': [engine_size],
            'Fuel Type': [fuel_type]
        })

        # Vorhersage des Preises für die neuen Daten
        predicted_price = trained_model.predict(new_data)

        print(f"Predicted Price for the {make} {model_name}, {year}: ${predicted_price[0]:,.2f}")
```
